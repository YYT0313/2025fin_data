```python
RSI = talib.RSI(Close, timeperiod=14)
```

```python
RSI_df = pd.DataFrame(RSI, columns=['RSI'], index=df.index)
```

```python
RSI_df
```

```python
buy_list = []
sell_list = []
```

```python
for index, row in RSI_df.iterrows():
    if pd.notna(row['RSI']):
        if row['RSI'] > 70:
            sell_list.append(row)
        elif row['RSI'] < 30:
            buy_list.append(row)
    else:
        continue
```

```python
sell_df = pd.DataFrame(sell_list)
buy_df = pd.DataFrame(buy_list)
```
