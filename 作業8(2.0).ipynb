```python
import numpy as np
import scipy.stats as si
import matplotlib.pyplot as plt
```

```python
S, K, r, sigma = 90, 100, 0.01, 0.3
```

```python
T_list = [1, 0.9, 0.5, 0.3, 0.1]
d1_list = []
d2_list = []
call_list = []
```

```python
for i in T_list:
    d1 = (np.log(S/K)+(r+0.5*sigma**2)*i)/(sigma*np.sqrt(i))
    d1_list.append(d1)
    d2 = d1 - sigma*np.sqrt(i)
    d2_list.append(d2)
    call = S*si.norm.cdf(d1,0.0,1.0)-K*np.exp(-r*i)*si.norm.cdf(d2,0.0,1.0)
    call_list.append(call)

print(d1_list,end='\n')

print(d2_list)

print(call_list)
```

```python
plt.plot(T_list, call_list)
```
